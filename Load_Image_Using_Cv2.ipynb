Load image using openCV CV2 - no need to convert my image in the numpy array

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
import requests
from io import BytesIO

# import cv2
from google.colab.patches import cv2_imshow
```

```python
url = "https://images.pexels.com/photos/3729464/pexels-photo-3729464.jpeg?auto=compress&cs=tinysrgb&w=600"

# Download the image using requests
response = requests.get(url)

# Read the image using OpenCV
img_array = np.asarray(bytearray(response.content), dtype=np.uint8)
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

# Display the image in Colab
cv2_imshow(img)
```

```python
# Reason of color differences is that:
# -------------------------------------------------------
# Matplotlib - RGB -> Red, Green, Blue
# openCV RGB -> Blue, Green, Red
```

```python
# Convert images into the right color of rgb

img_new = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_new)
```
