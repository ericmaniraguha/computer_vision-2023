RDGB (Red Green Blue)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
import requests
from io import BytesIO
```

```python
url = "https://images.pexels.com/photos/3433366/pexels-photo-3433366.jpeg?auto=compress&cs=tinysrgb&w=600"
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# Now you can work with the 'img' object
img
```

Convert the image in the numpy array

```python
img_arr = np.array(img)
type(img_arr)
```
